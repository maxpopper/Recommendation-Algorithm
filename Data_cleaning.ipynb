```python
import pandas as pd

# Load First TSV
title_basics_df = pd.read_csv("IMDB_data_TSVs/title.basics.tsv", sep='\t')
```

```python
# Step 3: Filter Data
# Filter Data to include only 'movie' type
title_basics_df_filtered = title_basics_df[title_basics_df['titleType'].isin(['movie'])]
```

```python
# Drop the "primaryTitle" and "endYear" columns
title_basics_df_filtered = title_basics_df_filtered.drop(columns=['primaryTitle', 'endYear'])

# Convert "startYear" column to integer
title_basics_df_filtered['startYear'] = pd.to_numeric(title_basics_df_filtered['startYear'], errors='coerce')

# Convert back to string and remove decimal
title_basics_df_filtered['startYear'] = title_basics_df_filtered['startYear'].astype('Int64').astype(str)
```

```python
# Load next TSV
title_akas_df = pd.read_csv("IMDB_data_TSVs/title.akas.tsv", sep='\t')
```

```python
# Select only the "language" column from title_akas_df
language_column = title_akas_df[['language']]

# Append the "language" column to title_basics_df_filtered based on "tconst"
title_basics_with_language = pd.concat([title_basics_df_filtered, language_column], axis=1)

# Print the first few rows of the merged DataFrame
title_basics_with_language.head()
```

```python
# Filter the DataFrame to include only titles with language 'en'
english_titles_df = title_basics_with_language[title_basics_with_language['language'] == 'en']

# Print the first few rows of the filtered DataFrame
english_titles_df.head()
```

```python
# Save the DataFrame with the appended "language" column to the same file
# title_basics_with_language.to_csv("IMDB_data_TSVs/cleaned_title_basics.tsv", sep='\t', index=False)
```

```python
# Load title.principals.tsv
# title_principals_df = pd.read_csv("IMDB_data_TSVs/title.principals.tsv", sep='\t')

# Merge title_basics_with_language with title.principals_df on "tconst"
merged_df = pd.merge(title_basics_with_language, title_principals_df[['tconst', 'nconst', 'category', 'job', 'characters']], on='tconst', how='left')

# Print the first few rows of the merged DataFrame
print(merged_df.head())
```

```python
# Drop the "job" column
merged_df = merged_df.drop(columns=['job'])

# Print the first few rows of the DataFrame to verify
merged_df.head()
```

```python
# Filter the DataFrame to include only 'actress', 'actor', and 'director' categories
filtered_df = merged_df[merged_df['category'].isin(['actress', 'actor', 'director'])]

# Print the first few rows of the filtered DataFrame
filtered_df.head()
```

```python
# Load title.ratings.tsv
title_ratings_df = pd.read_csv("IMDB_data_TSVs/title.ratings.tsv", sep='\t')

# Merge the main DataFrame with title_ratings_df on "tconst"
merged_df = pd.merge(filtered_df, title_ratings_df[['tconst', 'averageRating', 'numVotes']], on='tconst', how='left')

# Print the first few rows of the merged DataFrame
(merged_df.head()
```

```python
# Load name.basics.tsv
name_basics_df = pd.read_csv("IMDB_data_TSVs/name.basics.tsv", sep='\t')

# Merge the main DataFrame with name_basics_df on "nconst"
merged_df = pd.merge(merged_df, name_basics_df[['nconst', 'primaryName', 'knownForTitles']], on='nconst', how='left')

# Print the first few rows of the merged DataFrame
print(merged_df.head())
```

```python
merged_df.head()
```

```python
# Convert 'startYear' to string and filter movies after 1970
merged_df['startYear'] = merged_df['startYear'].astype(str)
filtered_df = merged_df[merged_df['startYear'] >= '1970']

# Filter 'language' to include only 'en'
filtered_df = filtered_df[filtered_df['language'] == 'en']

# Filter 'category' to include only 'actress', 'actor', and 'director'
filtered_df = filtered_df[filtered_df['category'].isin(['actress', 'actor', 'director'])]

# Filter 'isAdult' to include only '0'
filtered_df = filtered_df[filtered_df['isAdult'] == 0]

# Print the number of rows after filtering
print("Number of rows after filtering:", len(filtered_df))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv("data_for_ml.csv")

```

```python
# List of genres (excluding non-genre columns)
genres = ['Animation', 'Adventure', 'Mystery', 'Drama', 'Action', 'Romance', 'Musical', 
          'Comedy', 'Thriller', 'Crime', 'Documentary', 'Children', 'Sci-Fi', 'IMAX', 
          'Western', 'Fantasy', 'Horror']

# Calculate the average rating for each genre
genre_avg_ratings = df[genres].mean()

# Create a bar chart for average ratings by genre
plt.figure(figsize=(12, 8))
plt.bar(genres, genre_avg_ratings)
plt.xlabel('Genre')
plt.ylabel('Average Rating')
plt.title('Average Rating by Genre')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
